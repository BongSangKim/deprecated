## Review for basis change

Let $P_2$ be the vector space of degree 2 polynomials.
Let $T:P_2\to P_2$ be the map defined by $T(f(x)) = \frac{df}{dx}(x)$.

1. Show thaat $T$ is a linear transfomation

### Basis change

Let $\mathcal B_1 = \{1,x,x^2\}$ and $\mathcal B_2=\{1,(x-1),(x-1)^2\}$. Write $[T]_{\mathcal B_2}^{\mathcal B_2}$.

2. Show that $\mathcal B_1$ and $\mathcal B_2$ are basis of $P_2$.

3. Write $[T]_{\mathcal B_1}^{\mathcal B_1}$.

4. Write $[Id]_{\mathcal B_2}^{\mathcal B_1}$ and confirm that
    $$[T]_{\mathcal B_2}^{\mathcal B_2} = [Id]_{\mathcal B_1}^{\mathcal B_2} \cdot[T]_{\mathcal B_1}^{\mathcal B_1} \cdot [Id]_{\mathcal B_2}^{\mathcal B_1}$$

### Diagonalization

Define the inner product
    $$\langle f(x),g(x) \rangle = \int_0^1f(x)g(x)dx$$

5. Apply Gram-Schmidt on $\mathcal B_1$ to find an orthonormal basis $\{1,f_1(x),f_2(x)\}$ where $\deg f_n = n$.

6. Suppose $[f(x)]_{\mathcal B_1} = \begin{bmatrix} a_0 \\ a_1 \\ a_2 \end{bmatrix}$ and $[g(x)]_{\mathcal B_1} = \begin{bmatrix} b_0 \\ b_1 \\ b_2 \end{bmatrix}$. Compute $\langle f(x),g(x)\rangle$.

7. Find a symmetric matrix $S$ such that 
    $$\langle f(x),g(x)\rangle = [f(x)]_{\mathcal B_1}^T \cdot S \cdot [g(x)]_{\mathcal B_1}$$
    
8. Diagonalize $S = MDM^{-1}$ and explain the meaning of the matrix $M$.

### Schur and Jordan

9. Triangularize $[T]_{\mathcal B_1}^{\mathcal B_1}$ using an orthonormal basis.

10. Find Jordan form of $[T]_{\mathcal B_1}^{\mathcal B_1}$.

### SVD

Let $T:P_2\to P_1$ be the linear transformation $T(f(x)) = f'(x)$. Write $\mathcal B_1,\mathcal B_2$ for the basis of $P_1$ of the same type.

11. FInd SVD of $[T]_{\mathcal B_1}^{\mathcal B_1}$ and $[T]_{\mathcal B_1}^{\mathcal B_2}$
