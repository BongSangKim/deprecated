## 텍스트 데이터를 이용한 학습

```python
import gensim 

dante = []
with open("1.txt", encoding='utf-8') as file:
    for j, line in enumerate(file):
        dante.append(gensim.utils.simple_preprocess(line))

dante_model = gensim.models.Word2Vec (dante, size=10, window=10, min_count=2, workers=4)
dante_model.train(dante, total_examples=len(dante), epochs=1000)
```

```python
dante_model.wv.most_similar(positive=['roles'], topn=20)
```

## 트위터 API를 이용한 데이터 마이닝

```python
'''
twitter 크롤링 
[출처] #파이썬. Python으로 트위터 크롤링(twitter crawling)하기|작성자 안재형
'''

import tweepy

# 트위터 앱의 Keys and Access Tokens 탭 참조(자신의 설정 값을 넣어준다)
consumer_key = ".."
consumer_secret = ".."

# 1. 인증요청(1차) : 개인 앱 정보 
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)

access_token = ".."
access_token_secret= ".."

# 2. access 토큰 요청(2차) - 인증요청 참조변수 이용
auth.set_access_token(access_token, access_token_secret)

# 3. twitter API 생성  
api = tweepy.API(auth)

keyword = "국정농단";     # 자신이 검색하고 싶은 키워드 입력 
search = [] # 크롤링 결과 저장할 변수   

cnt = 1
while(cnt <= 10):   # 10page 대상으로 크롤링
   tweets = api.search(keyword)
   for tweet in tweets:
       search.append(tweet)
   cnt += 1

print(len(search)) # 문서 길이 
print(search[0]) # 첫번째 text 보기 
```

```python
for tweet in tweepy.Cursor(api.search, q=('"국정농단"'), since='2014-09-16', until='2019-05-15').items(100):
    print("Name:", tweet.author.name)
    print("Screen-name:", tweet.author.screen_name)
    print("Tweet created:", tweet.created_at)
    print("Tweet:", tweet.text)
    print("Retweeted:", tweet.retweeted)
    print("Favourited:", tweet.favorited)
    print("Location:", tweet.user.location)
    print("Time-zone:", tweet.user.time_zone)
    print("Geo:", tweet.geo)
    print("//////////////////")
    print("Hashtags:", tweet.entities.get('hashtags'))
```

```python
tweets = tweepy.Cursor(api.search, q=('"국정농단"'), since='2014-09-16', until='2019-05-15', tweet_mode='extended').items(10)
for tweet in tweets:
    print("Tweet:", tweet.full_text)
```
