* 노래 가사는 문장 구분 없이 학습시킬 것.
* simple_preprocess는 I와 같은 간단한 단어를 제외시킴. 직접 전처리 해야함.

```python
import gensim 

beetles = ""
with open("beetles.txt", encoding='utf-8') as file:
    for j, line in enumerate(file):
        line = line.replace("\n", " ")
        beetles += line
print(beetles)
```

```python
beetles = gensim.utils.simple_preprocess(beetles)
print(beetles)
```

```python
import gensim 

beetles = []
with open("beetles.txt", encoding='utf-8') as file:
    temp = [""]
    for j, line in enumerate(file):
        words = line.replace("\n", " ").replace(",", " ").replace(".", " ").split()
        for word in words:
            if "'" in word:
                idx = word.index("'")
                temp.append(word[:idx])
            else:
                temp.append(word)
    beetles.append(temp)
```

```python
print(beetles)
```

```python
beetles_model = gensim.models.Word2Vec (beetles, size=10, window=10, min_count=2, workers=4)
beetles_model.train(beetles, total_examples=len(beetles), epochs=500)
```

```python
wordlist = []
count = []
for word, obj in beetles_model.wv.vocab.items():
    wordlist.append(word)
    count.append(obj.count)
for _ in range(100):
    idx = count.index(max(count))
    print(wordlist[idx], max(count))
    del count[idx]
    del wordlist[idx]
```

```python
import gensim 

queen = []
with open("queen.txt") as file:
    for j, line in enumerate(file):
        queen.append(gensim.utils.simple_preprocess(line))

queen_model = gensim.models.Word2Vec (queen, size=10, window=10, min_count=2, workers=4)
queen_model.train(queen, total_examples=len(queen), epochs=1000)
```

```python
wordlist = []
count = []
for word, obj in queen_model.wv.vocab.items():
    wordlist.append(word)
    count.append(obj.count)
for _ in range(100):
    idx = count.index(max(count))
    print(wordlist[idx], max(count))
    del count[idx]
    del wordlist[idx]
```

```python
beetles_model.wv.most_similar(positive=['love'], topn=20)
```

```python
queen_model.wv.most_similar(positive=['you'], topn=20)
```

```python
import numpy as np

v1 = 'love'
v2 = 'me'

w = beetles_model.wv['man']
w = w / np.linalg.norm(w)
italia_1 = beetles_model.wv[v1] 
italia_1 = italia_1 / np.linalg.norm(italia_1)
italia_2 = beetles_model.wv[v2]
italia_2 = italia_2 / np.linalg.norm(italia_2)

print(np.dot(w,italia_1), np.dot(w,italia_2))

w = queen_model.wv['man']
w = w / np.linalg.norm(w)
afgan_1 = queen_model.wv[v1] 
afgan_1 = afgan_1 / np.linalg.norm(afgan_1)
afgan_2 = queen_model.wv[v2]
afgan_2 = afgan_2 / np.linalg.norm(afgan_2)

print(np.dot(w,afgan_1), np.dot(w,afgan_2))
```
