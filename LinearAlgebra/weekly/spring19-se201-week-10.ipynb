## Lecture 18

### Review

* Difference equation in general
  * stable if all e.vals has abs $<1$: $u_k\to0$ eventually.
  * neutrally stable if some e.vals have abs $=1$, and all the other has $<1$.
    * Example: $A = \begin{bmatrix} .2 & .4 & .3 \\ .4 & .2 & .3 \\ .4 & .4 & .4 \end{bmatrix}$
    * Prop: Markov matrix always has $1$ as an e.val.
    * E.vec for e.val $1$ is called the steady-state solution.
  * unstable if some e.vals have abs $>1$.
    * stable: $\begin{bmatrix}0 & 4 \\ 0 & 1/2\end{bmatrix}$
* Leontif's intput-output model
  * $p$: production, $y$: demand. $A = (a_{ij})$, $a_{ij}$: amount of product $j$ need to produce $1$ unit of product $i$.
  * $y = p - Ap$. $p = (I-A)^{-1}y$.
  * Let  $\lambda_1$ be the largest e.val of $A$.
    * If $\lambda>1$, then $(I-A)^{-1}$ has a negative e.val. 
      * Example: $A = \begin{bmatrix}0 & 2 \\2 & 0\end{bmatrix}$. The economy may not produce some products.
    * If $\lambda=0$, then $(I-A)^{-1}$ does not exist.
    * If $\lambda<1$, then $(I-A)^{-1}$ all e.val are nonnegative. 
      * Example: $A = \begin{bmatrix}.5 & 2 \\0 & .5\end{bmatrix}$
    * Example: $A = \begin{bmatrix} .4 & .0 & .1 \\ .0 & .1 & .8 \\ .5 & .7 .1 \end{bmatrix}$. char. poly: $x^3 - 3/5x^2 - 13/25x + 9/40 = (x-0.9)(x^2-3/10x-1/4) = 0$. Max e.val = $.9$. The economy can produce anything.

  * $u = c_1e^{\lambda_1t}x_1 + c_2e^{\lambda_2t}x_2 = S\textrm{diag}(\lambda_1,\lambda_2)\begin{bmatrix}x_1\\x_2\end{bmatrix}$
  * The equation $ u' = Au$ has a solution $u = e^{At}u_0$.

  * If $A = S\Lambda S^{-1}$, $S = \textrm{diag}(\lambda_1,\cdots,\lambda_n)$, then
    $$u = c_1e^{\lambda_1t}x_1 + \cdots + c_ne^{\lambda_n t}x_n$$
    where $x_1,\cdots, x_n$ are e.vec for e.vals.

* Stability:
  * If $\textrm{Re}\lambda_i<0$ for all $i$, then **stable**.
  * If some $\textrm{Re}\lambda_i=0$, then **neutrally stable**.
  * If any $\textrm{Re}\lambda_i>0$, then **unstable**.

* Examples
  * $A = \begin{bmatrix}-2 & 1\\1 & -2\end{bmatrix}$: stable
  * $A = \begin{bmatrix}0 & -1\\1 & 0\end{bmatrix}$: neutrally stable
  * Continuous Markov process
    * $A = \begin{bmatrix} -1 & 1 \\ 1 & -1 \end{bmatrix}$
    * sum of each column is $0=\lambda_{\textrm{max}}$
